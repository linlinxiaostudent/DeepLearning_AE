```python
import numpy as np
from sklearn.svm import SVC
rng = np.random.RandomState(0)
X = rng.rand(100, 10)
print(X)
y = rng.binomial(1, 0.5, 100)
print(y)
X_test = rng.rand(5, 10)
print(X_test)
```

```python
clf=SVC()
clf.set_params(kernel='linear').fit(X,y)
```

```python
clf.predict(X_test)
```

在这里，默认内核rbf首先被改变到linear估计器被构造之后SVC()，并且改回到rbf重新设计估计器并进行第二预测。

**多类与多标签拟合**

```python
from sklearn.svm import SVC
from sklearn.multiclass import OneVsRestClassifier
from sklearn.preprocessing import LabelBinarizer
X = [[1, 2], [2, 4], [4, 5], [3, 2], [3, 1]]
y = [0, 0, 1, 1, 2]
classif = OneVsRestClassifier(estimator=SVC(random_state=0))
classif.fit(X, y).predict(X)
```

```python
y = LabelBinarizer().fit_transform(y)
classif.fit(X, y).predict(X)
```

**线性回归**

```python
from sklearn import linear_model
reg = linear_model.LinearRegression()
reg.fit([[0,0],[1,1],[2,2]],[0,1,2])
```

```python
reg.coef_
```

LinearRegression ( 线性回归 ) 将采用其拟合方法数组 X ， y 并将其线性模型的系数 w 存储在其 coef_ 成员中

**岭回归 Ridge Regression**
岭回归通过对系数的大小施加惩罚项来解决普通话最小二乘的一些问题，ridge coefficients ( 岭系数 ) 最小化了惩罚的残差平方和。

```python
from sklearn import linear_model
reg = linear_model.Ridge(alpha=0.5)
reg.fit([[0,0],[1,1],[2,2]],[0,1,2])
```

```python
reg.coef_
```

```python
reg.intercept_
```

reg.coef_：系数w

reg.intercept_:截距

**岭回归设置正则化参数：通过广义交叉验证**

RidgeCV通过设置Alpha参数设置交叉验证来实现岭回归，该对象的工作方式和GridSearchCV相同，只是它默认为 Generalized Cross-Validation ( 通用交叉验证 ) （GCV），这是一种有效的一对一交叉验证形式。

```python
from sklearn import linear_model
reg = linear_model.RidgeCV(alphas=[0.1,1.0,10.0])
reg.fit([[0,0],[1,1],[2,2]],[0,0.1,1])
```

```python
reg.alpha_
```

**Lasso**

Lasso是估计稀疏系数的线性模型，它在一些情况下是有用的，因为他倾向于使用较少参数值的解决方案，有效地减少给定解决方案所依赖的变量的数量。为此，Lasso及其变体是compressed sensing（压缩感测领域）的基础，在某些条件下，它可以恢复精确的非零权重集。

```python
from sklearn import linear_model
reg = linear_model.Lasso(alpha=0.1)
reg.fit([[0,0],[1,1]],[0,1])
```

```python
reg.predict([[1,1]])
```
